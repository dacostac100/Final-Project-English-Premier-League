## Data Analysis
### Who are the top 10 players based on goals scored?

```python
top_10_players = df
top_10_players["Goals"] = pd.to_numeric(top_10_players["Goals"], errors="coerce")
top_10_players = top_10_players.nlargest(10, "Goals")[["Player", "Goals", "Squad"]]
plt.figure(figsize=(10, 6))
plt.barh(top_10_players["Player"], top_10_players["Goals"], color="green")
plt.xlabel("Goals")
plt.ylabel("Player")
plt.title("Top 10 EPL Players")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()  
for i, (goals, Squad) in enumerate(zip(top_10_players["Goals"], top_10_players["Squad"])):
    plt.text(goals + 1, i, Squad, va='center', fontsize=10, color='blue')  # Add text slightly to the right of the bars

plt.tight_layout()
plt.show() 
```

###  Comparison of actual goals vs expected goals for top 10 players.

```python
import numpy as np
actual_expected = df
actual_expected['Goals'] = pd.to_numeric(df['Goals'], errors='coerce')
actual_expected['Expected_goals'] = pd.to_numeric(df['Expected_goals'], errors='coerce')

actual_expected = actual_expected.dropna(subset=['Goals', 'Expected_goals'])

top_players = actual_expected.nlargest(10, 'Goals')[['Player', 'Squad', 'Goals', 'Expected_goals']]

top_players.set_index('Player', inplace=True)

ind = np.arange(len(top_players))  
width = 0.35  

fig, ax = plt.subplots(figsize=(12, 7))

actual_goals = ax.bar(ind - width/2, top_players['Goals'], width, label='Actual Goals', color='skyblue')

expected_goals = ax.bar(ind + width/2, top_players['Expected_goals'], width, label='Expected Goals', color='lightgreen')

ax.set_xlabel('Player')
ax.set_ylabel('Number of Goals')
ax.set_title('Actual Goals vs Expected Goals for Top 10 Players')
ax.set_xticks(ind)
ax.set_xticklabels(top_players.index, rotation=45)
ax.legend()

for i in range(len(top_players)):
    ax.text(i - width/2, top_players['Goals'].iloc[i] + 0.5, top_players['Squad'][i], ha='center', va='bottom', fontsize=9, rotation=90)
    ax.text(i + width/2, top_players['Expected_goals'].iloc[i] + 0.5, top_players['Squad'][i], ha='center', va='bottom', fontsize=9, rotation=90)

plt.tight_layout()
plt.show()
```

### Did younger more physical/less experience player (age 17-25) scored more goals than older less physical/more experience player (age 26-39)?

```python

younger_older = df
younger_older['Goals'] = pd.to_numeric(younger_older['Goals'], errors='coerce')
younger_older['Age'] = pd.to_numeric(younger_older['Age'], errors='coerce')

younger_older['Age_Group'] = pd.cut(younger_older['Age'], bins=[16, 25, 39], labels=['17-25 (Younger)', '26-39 (Older)'])

average_goals_by_age = younger_older.groupby('Age_Group')['Goals'].mean()

plt.figure(figsize=(8, 6))
average_goals_by_age.plot(kind='bar', color=['skyblue', 'lightgreen'])

plt.title('Average Goals Scored by Age Group')
plt.ylabel('Average Goals')
plt.xlabel('Age Group')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
